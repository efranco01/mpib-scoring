```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
# Define the cosine squared function
def cos2(x, a, b, c, d):
    return a * np.cos(b * x + c)**2 + d
# Define the dataset
y_data = np.array([74, 72, 64, 53, 38, 24, 11, 3, -1, -1, -1, 7, 18, 31, 45, 56, 66, 72, 75])
x_data = np.linspace(0, np.pi, len(y_data))
# Fit the dataset to the cosine squared function
popt, pcov = curve_fit(cos2, x_data, y_data)
# Print the parameters of the fitted function
print("a = %f" % popt[0])
print("b = %f" % popt[1])
print("c = %f" % popt[2])
print("d = %f" % popt[3])
# Plot the original dataset and the fitted function
plt.plot(x_data, y_data, 'b+', label='data')
plt.plot(x_data, cos2(x_data, *popt), 'r-', label='fit')
plt.legend()
plt.show()
```
